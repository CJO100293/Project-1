```python
import os
import csv
import pandas as pd
import numpy as np

#open and read the file
csvpath = os.path.join('Resources', 'NPS-Unit-List.csv')

dataList = []
#receive the data from file to list
with open(csvpath) as csvopen:
    dataFile = csv.reader(csvopen, delimiter=',')
   
    #header is excluded but stored as a list
    header = next(dataFile)
    Anotherheader = next(dataFile)
    #from data from file to a 2d list
    for row in dataFile:
        dataList.append(row)

#don't need the file anymore
csvopen.close() 

print(dataList)
```

```python
national_parks = []
parkCodes = []

#add some parks that weren't included in the .csv file using the NPS website
national_parks.append("American Memorial Park")
parkCodes.append("amme")
national_parks.append("Ala Kahakai National Historic Trail")
parkCodes.append("alka")
national_parks.append("Anacostia Park")
parkCodes.append("anac")
national_parks.append("Jamestown")
parkCodes.append("jame")
national_parks.append("Yosemite")
parkCodes.append("yose")
national_parks.append("Yellowstone")
parkCodes.append("yell")
national_parks.append("Obed Wild & Scenic River")
parkCodes.append("obed")
national_parks.append("Zion National Park")
parkCodes.append("zion")
national_parks.append("Alagnak Wild River")
parkCodes.append("alag")
national_parks.append("Yorktown Battlefield")
parkCodes.append("york")

for park in dataList:
    
    #clean up the national park name
    name = park[0].title()
    name = name.split("(")[0]
    name = name.split("-")[0]
    name = name.split("  ")[0]
    name = name.split("_")[0]
    national_parks.append(name.strip())
    
    #clean up park code (only keep first 4 letters)
    code = park[2].lower()
    parkCodes.append(code[:4])
```

```python
df = pd.DataFrame({"Full Name" : national_parks, 
                   "Park Code" : parkCodes})

df = df.head(427)
df = df.set_index("Full Name")
df.to_csv("output_data/park_dict_df.csv")
df
```
